```python
# important library

import cv2
from matplotlib import pyplot as plt
import numpy as np

I = cv2.imread("test.png")
I = cv2.cvtColor(src = I, code = cv2.COLOR_RGB2BGR)

plt.imshow(I)
plt.show()
```

```python
# Reduce the noise by Gaussian Blur
G = cv2.cvtColor(src = I, code = cv2.COLOR_BGR2GRAY)
I = cv2.GaussianBlur(G,(5,5),0)

plt.imshow(I)
plt.set_cmap('gray')

plt.show()
```

```python
B = cv2.adaptiveThreshold(G, maxValue = 255,
 adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
  thresholdType = cv2.THRESH_BINARY,
   blockSize = 25,C = 5)

plt.imshow(B)
plt.show()
```

```python
low_pass = np.array([[1,4,1], [4,7,4], [1,4,1]], dtype=float) /27

L = cv2.filter2D(src = B, ddepth = -1, kernel = low_pass)

plt.imshow(L)
plt.show()
```

```python
sharpen = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]], dtype = float)
S = cv2.filter2D(src = L, ddepth = -1, kernel = sharpen)


plt.imshow(S)
plt.show()
```

```python
edges = cv2.Canny(G,10,200)
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
I = cv2.imread('test.png')
# Apply bilateral filter
bilateral = cv2.bilateralFilter(I, 15, 75, 75)

plt.imshow(cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blur = cv2.GaussianBlur(I,(5,5),0)

# This function is use for drawing the edges onto the gaussianblur-filtered image
def draw_edge(img, edge):
    for i in range(len(edge)):
        for j in range (len(edge[0])):
            if edge[i][j] == 0:
                img[i][j] = [0,0,0]
    return img
img_output = draw_edge(blur,S)

plt.imshow(cv2.cvtColor(img_output,cv2.COLOR_BGR2RGB))
plt.show()
```
